```python
import ast
def extract(obj):
    name = []
    for i in ast.literal_eval(obj):
        name.append(i["name"])
    return name
```

```python
movies["genres"] = movies["genres"].apply(extract)
```

```python
movies["keywords"] = movies["keywords"].apply(extract)
```

```python
def cast(obj):
    cast = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            cast.append(i["name"])
        count+=1
        
    return cast
```

```python
movies["cast"] = movies["cast"].apply(cast)
```

```python
import ast
def extract(obj):
    name = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            name.append(i["name"])
            break
    return name
```

```python
movies["crew"] = movies["crew"].apply(extract)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" " , "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" " , "") for i in x])
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" " , "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x : [i.replace(" " , "") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.sample()
```

```python
movies = movies[["id","title","tags"]]
```

```python
movies.sample()
```

```python
movies["tags"] = movies["tags"].apply(lambda x : " ".join(x))
```

```python
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x : x.lower())
movies.head(1)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def steam(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies["tags"] = movies["tags"].apply(steam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector = cv.fit_transform(movies["tags"]).toarray()
```

```python
cv.get_feature_names_out()[50:1000]
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[0]
```

```python
movies[movies["title"] == "Batman Begins"].index[0]
```

```python
def recommend(movie):
    movies_index =movies[movies["title"] == movie].index[0]
    distance = similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
